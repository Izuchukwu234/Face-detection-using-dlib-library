## Face Detection using dlib library

#### Importing the necessary libraries

```python
import cv2
import dlib
import matplotlib.pyplot as plt
```

#### Loading and displaying the image

```python
raw_image = cv2.imread("prevention-image-of-large-group-smiling-300x200.jpg", cv2.IMREAD_COLOR)

gray_img = cv2.cvtColor(raw_image, cv2.COLOR_BGR2GRAY)
```

```python
# Displaying the original image
plt.figure(figsize = (9, 9))
plt.imshow(cv2.cvtColor(raw_image, cv2.COLOR_BGR2RGB))
plt.title('Original Picture')
plt.axis("off")
plt.show()
```

#### Creating a face detector class

```python
face_detector = dlib.get_frontal_face_detector()
faces = face_detector(gray_img)
```

#### Making the facial detection identifier and displaying the processed image

```python
for i in faces:
    x1, y1, x2, y2 = i.left(), i.top(), i.right(), i.bottom()
    cv2.rectangle(raw_image, (x1, y1), (x2, y2), (0, 255, 0), 4)
```

```python
# Displaying the image with detection
plt.figure(figsize = (9, 9))
plt.imshow(cv2.cvtColor(raw_image, cv2.COLOR_BGR2RGB))
plt.title("Picture with detected faces")
plt.axis("off")
plt.show()
```
